```python
import pandas as pd
from pycaret.time_series import *
```

```python
data = pd.read_csv('/content/Daily Electricity price.csv')
```

```python
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')
data.set_index('date', inplace=True)
print(data.head())
```

```python
setup(data=data['demand'], fh=7, seasonal_period=7)
```

```python
arima_model = create_model('arima')
```

```python
plot_model(arima_model, plot='forecast')
```

```python
final_arima_model = finalize_model(arima_model)
```

```python
forecast = predict_model(final_arima_model, fh=7)
print(forecast)
```
