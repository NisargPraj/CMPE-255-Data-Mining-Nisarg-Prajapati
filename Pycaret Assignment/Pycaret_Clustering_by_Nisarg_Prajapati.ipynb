```python
import pandas as pd
from pycaret.clustering import *
```

```python
data = pd.read_csv('/content/forest.csv')

data.head()
```

```python
cluster = setup(data, normalize=True, use_gpu=True)
```

```python
k_means = create_model('kmeans')

clustered_data = assign_model(k_means)
clustered_data.head()
```

```python
plot_model(k_means, plot='cluster')
```
